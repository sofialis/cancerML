```python
def init_data_UCEC(np_data):
    # age 
    age = np_data[:,2].astype(int)

    # clinical_stage
    clinical_stage = np_data[:,4]    
    clinical_stage_1 = np.where(clinical_stage=="Stage I", 1, 0)
    clinical_stage_1a = np.where(clinical_stage=="Stage IA", 1, 0)
    clinical_stage_1b = np.where(clinical_stage=="Stage IB", 1, 0)
    clinical_stage_1c = np.where(clinical_stage=="Stage IC", 1, 0)
    clinical_stage_2 = np.where(clinical_stage=="Stage II", 1, 0)
    clinical_stage_2a = np.where(clinical_stage=="Stage IIA", 1, 0)
    clinical_stage_2b = np.where(clinical_stage=="Stage IIB", 1, 0)
    clinical_stage_3 = np.where(clinical_stage=="Stage III", 1, 0)
    clinical_stage_3a = np.where(clinical_stage=="Stage IIIA", 1, 0)
    clinical_stage_3b = np.where(clinical_stage=="Stage IIIB", 1, 0)
    clinical_stage_3c = np.where(clinical_stage=="Stage IIIC", 1, 0)
    clinical_stage_3c1 = np.where(clinical_stage=="Stage IIIC1", 1, 0)
    clinical_stage_3c2 = np.where(clinical_stage=="Stage IIIC2", 1, 0)
    clinical_stage_4 = np.where(clinical_stage=="Stage IV", 1, 0)
    clinical_stage_4a = np.where(clinical_stage=="Stage IVA", 1, 0)
    clinical_stage_4b = np.where(clinical_stage=="Stage IVB", 1, 0)
    
    # histological_type
    histological_type = np_data[:,5]    
    histological_type_endometrioid = np.where(histological_type=="Endometrioid endometrial adenocarcinoma", 1, 0)
    histological_type_mixed_serous_endo = np.where(histological_type=="Mixed serous and endometrioid", 1, 0)
    histological_type_serous = np.where(histological_type=="Serous endometrial adenocarcinoma", 1, 0)

    # histological_grade
    histological_grade = np_data[:,6] 
    histological_grade_g1 = np.where(histological_grade=="G1", 1, 0)
    histological_grade_g2 = np.where(histological_grade=="G2", 1, 0)
    histological_grade_g3 = np.where(histological_grade=="G3", 1, 0)
    histological_grade_high = np.where(histological_grade=="High Grade", 1, 0)

    # initial_pathologic_dx_year
    initial_pathologic_dx_year = np_data[:,7].astype(int)

    # menopause_status

    # vital_status
    vital_status = np_data[:,9]
    vital_status = np.array([l == 'Alive' for l in vital_status]).astype('int') #0 Dead, 1 Alive

    # tumor_status
    tumor_status = np_data[:,10] 
    tumor_status = np.array([l == 'WITH TUMOR' for l in tumor_status]).astype('int') #0 no tum, 1 with tum

    # death_days_to
    death_days_to = np_data[:, 11] 

    # new_tumor_event_type
    new_tumor_event_type = np_data[:,12] # N/A - 0, other - 1
    new_tumor_event_type = np.array([l != '#N/A' for l in new_tumor_event_type]).astype('int') #0 no tum, 1 with tum

    # new_tumor_event_site

    #new_tumor_event_site_other

    #new_tumor_event_dx_days_to
    new_tumor_event_dx_days_to = np_data[:,15]
    new_tumor_event_dx_days_to = np.where(new_tumor_event_dx_days_to=="#N/A", 0 , new_tumor_event_dx_days_to).astype(int)

    # treatment_outcome_first_course

    outcome = np_data[:,16];
    outcome = np.where(outcome=="Complete Remission/Response", 0, outcome)
    outcome = np.where(outcome=="Partial Remission/Response", 3, outcome)
    outcome = np.where(outcome=="Progressive Disease", 2, outcome)
    outcome = np.where(outcome=="Stable Disease", 1, outcome)
    outcome = np.where(outcome=="Persistent Disease", 4, outcome)

    m = np_data.shape[0]
    X = np.stack([np.ones(m), age, histological_grade_g1,histological_grade_g2,histological_grade_g3,histological_grade_high, histological_type_endometrioid,histological_type_mixed_serous_endo,histological_type_serous, clinical_stage_1, clinical_stage_1a, clinical_stage_1b, clinical_stage_1c, clinical_stage_2, clinical_stage_2a,clinical_stage_2b,clinical_stage_3,clinical_stage_3a,clinical_stage_3b,clinical_stage_3c,clinical_stage_3c1,clinical_stage_3c2,clinical_stage_4,clinical_stage_4a,clinical_stage_4b, initial_pathologic_dx_year, tumor_status, new_tumor_event_type, new_tumor_event_dx_days_to], axis=1).astype(int)
    feature_labels = ['1-vector', 'age', 'histological_grade_g1','histological_grade_g2','histological_grade_g3','histological_grade_high', 'histological_type_endometrioid','histological_type_mixed_serous_endo','histological_type_serous', 'clinical_stage_1', 'clinical_stage_1a', 'clinical_stage_1b', 'clinical_stage_1c', 'clinical_stage_2', 'clinical_stage_2a','clinical_stage_2b','clinical_stage_3','clinical_stage_3a','clinical_stage_3b','clinical_stage_3c','clinical_stage_3c1','clinical_stage_3c2','clinical_stage_4','clinical_stage_4a','clinical_stage_4b', 'initial_pathologic_dx_year', 'tumor_status', 'new_tumor_event_type', 'new_tumor_event_dx_days_to']
    return X, feature_labels, vital_status, death_days_to, outcome
```

```python
def init_data_HNSC(np_data):
    # age 
    age = np_data[:,2].astype(int)

    # gender
    gender = np_data[:,3]
    gender = np.array([l == 'FEMALE' for l in gender]).astype('int') #0 man, 1 woman

    # ajcc_pathologic_tumor_stage
    tumor_stage = np_data[:,4]
    tumor_stage_1 = np.where(tumor_stage=="Stage I", 1, 0)
    tumor_stage_2 = np.where(tumor_stage=="Stage II", 1, 0)
    tumor_stage_3 = np.where(tumor_stage=="Stage III", 1, 0)
    tumor_stage_4a = np.where(tumor_stage=="Stage IVA", 1, 0)
    tumor_stage_4b = np.where(tumor_stage=="Stage IVB", 1, 0)

    # clinical_stage
    clinical_stage = np_data[:,5]
    clinical_stage_1 = np.where(clinical_stage=="Stage I", 1, 0)
    clinical_stage_2 = np.where(clinical_stage=="Stage II", 1, 0)
    clinical_stage_3 = np.where(clinical_stage=="Stage III", 1, 0)
    clinical_stage_4a = np.where(clinical_stage=="Stage IVA", 1, 0)
    clinical_stage_4b = np.where(clinical_stage=="Stage IVB", 1, 0)
    clinical_stage_4c = np.where(clinical_stage=="Stage IVC", 1, 0)

    # histological_type
    histological_type = np_data[:,6] # ignore, they are all the same

    # histological_grade
    histological_grade = np_data[:,7] 
    histological_grade_g1 = np.where(histological_grade=="G1", 1, 0)
    histological_grade_g2 = np.where(histological_grade=="G2", 1, 0)
    histological_grade_g3 = np.where(histological_grade=="G3", 1, 0)
    histological_grade_g4 = np.where(histological_grade=="G4", 1, 0)
    histological_grade_gx = np.where(histological_grade=="GX", 1, 0)

    # initial_pathologic_dx_year
    initial_pathologic_dx_year = np_data[:,8].astype(int)

    # vital_status
    vital_status = np_data[:,9]
    vital_status = np.array([l == 'Alive' for l in vital_status]).astype('int') #0 Dead, 1 Alive

    # tumor_status
    tumor_status = np_data[:,10] 
    tumor_status = np.array([l == 'WITH TUMOR' for l in tumor_status]).astype('int') #0 no tum, 1 with tum

    # death_days_to
    death_days_to = np_data[:, 11] 

    # new_tumor_event_type
    new_tumor_event_type = np_data[:,12] # N/A - 0, other - 1
    new_tumor_event_type = np.array([l != '#N/A' for l in new_tumor_event_type]).astype('int') #0 no tum, 1 with tum

    # new_tumor_event_site
    new_tumor_event_site = np_data[:,13]
    new_tumor_event_site = np.array([l != '#N/A' for l in new_tumor_event_site]).astype('int') #0 no tum, 1 with tum

    #new_tumor_event_site_other

    #new_tumor_event_dx_days_to
    new_tumor_event_dx_days_to = np_data[:,15]
    new_tumor_event_dx_days_to = np.where(new_tumor_event_dx_days_to=="#N/A", 0 , new_tumor_event_dx_days_to).astype(int)

    # treatment_outcome_first_course

    outcome = np_data[:,16];
    outcome = np.where(outcome=="Complete Remission/Response", 0, outcome)
    outcome = np.where(outcome=="Partial Remission/Response", 3, outcome)
    outcome = np.where(outcome=="Progressive Disease", 2, outcome)
    outcome = np.where(outcome=="Stable Disease", 1, outcome)
    outcome = np.where(outcome=="Persistent Disease", 4, outcome)
    
    m = np_data.shape[0]
    X = np.stack([np.ones(m), age, gender, tumor_stage_1, tumor_stage_2, tumor_stage_3, tumor_stage_4a, tumor_stage_4b, histological_grade_g1, histological_grade_g2, histological_grade_g3, histological_grade_g4, histological_grade_gx, clinical_stage_1, clinical_stage_2, clinical_stage_3, clinical_stage_4a, clinical_stage_4b, clinical_stage_4c, initial_pathologic_dx_year, tumor_status, new_tumor_event_type, new_tumor_event_dx_days_to], axis=1).astype(int)
    feature_labels = ['1-vector', 'age', 'gender', 'tumor_stage_1', 'tumor_stage_2', 'tumor_stage_3', 'tumor_stage_4a', 'tumor_stage_4b', 'histological_grade_g1', 'histological_grade_g2', 'histological_grade_g3', 'histological_grade_g4', 'histological_grade_gx', 'clinical_stage_1', 'clinical_stage_2', 'clinical_stage_3', 'clinical_stage_4a', 'clinical_stage_4b', 'clinical_stage_4c', 'initial_pathologic_dx_year', 'tumor_status', 'new_tumor_event_type', 'new_tumor_event_dx_days_t']
    return X, feature_labels, vital_status, death_days_to, outcome
```
